```python
import pandas as pd
import random
import seaborn as sn
from matplotlib import pyplot as plt
import numpy as np
```

```python
def right_orientation():
    if random.randint(1, 6) >= 4:
        return True
    return False
```

```python
def simulation(rolls, sides):
    effective_collisions = 0
    activation_energy = []
    for i in range(2, (2 *sides) + 1):
        activation_energy.append(i)
    effective_collisions = [0] * ((sides * 2) - 1)
    maxwell_boltzman = [0] * ((sides * 2) - 1)
    
    for i in range(rolls):
            particle_energy1 = random.randint(1, sides) + random.randint(1, sides)
            particle_energy2 = random.randint(1, sides) + random.randint(1, sides)
            maxwell_boltzman[particle_energy1 - 2] += 1
            maxwell_boltzman[particle_energy2 - 2] += 1
            if not right_orientation(): continue
            for curr_activation in activation_energy:
                if particle_energy1 >= curr_activation and particle_energy2 >= curr_activation:
                    effective_collisions[curr_activation - 2] += 1
                else:
                    break
    return maxwell_boltzman, effective_collisions
```

# Low Temp

### 2000000 Particles

```python
data = simulation(1000000, 6)
x = [i for i in range(2, 13)]
```

```python
plt.plot(x, data[0])
```

```python
plt.bar(x, data[0])
```

```python
# activation energy vs effective collisions
plt.plot(x, data[1])
```
